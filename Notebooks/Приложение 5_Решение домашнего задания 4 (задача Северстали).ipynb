```python
# загружаем необходимые библиотеки
import pandas as pd
import numpy as np
```

```python
# считываем массив признаков и массив меток
# данные взяты отсюда https://drive.google.com/drive/folders/16kZDOe7zxrwdIW99OIE9Fc_UvfV9wiDD
clf_data = pd.read_csv('Data/X_Severstal.csv', sep=',')
clf_target = pd.read_csv('Data/y_Severstal.csv', sep=',')
```

```python
# смотрим на данные
clf_data.head()
```

```python
# смотрим стастистики, замечаем, что многие переменные
# имеют нормальное распределение (практически одинаковые
# значения средних и медиан)
clf_data.describe()
```

```python
# смотрим типы переменных и пропуски
clf_data.info()
```

```python
# выводим количество уникальных значений
for i in clf_data:
    print(i, clf_data[i].nunique())
```

```python
# предположим, что переменные с менее чем 10 уникальными значениями являются
# категориальными, и записываем в отдельный список
cat_var = clf_data.loc[:, clf_data.apply(pd.Series.nunique) < 10].columns
# меняем переменные в списке на категориальные и добавляем
# к названиям этих переменных префикс catvar_
for i in cat_var:
    clf_data[i] = clf_data[i].astype('object')    
    clf_data.rename(columns={f'{i}':f'catvar_{i}'}, inplace=True)

clf_data.info()
```

```python
# пересоздаем список категориальных переменных
cat_var = [c for c in clf_data.columns if clf_data[c].dtype.name == 'object']
# смотрим редкие категории
for i in cat_var:
     print(clf_data[i].value_counts(dropna=False))
```

```python
# заменяем пропуски на 9999
for i in cat_var:
    clf_data[i] = np.where(clf_data[i].isnull(), '9999', clf_data[i])
```

```python
# категории с менее 1000 наблюдений записываем в категорию Rare_value
for i in cat_var:
    clf_data.loc[clf_data[i].value_counts()[clf_data[i]].values < 1000, i] = 'Rare_value'
```

```python
# создаем список количественных переменных
num_columns = [c for c in clf_data.columns if clf_data[c].dtype.name != 'object']
for i in num_columns:
    # заменяем пропуски на 9999
    clf_data[i].fillna(9999, inplace=True)
    # выполняем downcasting для уменьшения размера данных
    clf_data[i] = clf_data[i].astype(np.float16)

clf_data.info()
```

```python
# разделим выборку на обучение и тест
from sklearn.model_selection import train_test_split
clf_train_data, clf_test_data, clf_train_labels, clf_test_labels = train_test_split(clf_data, clf_target,
                                                                                    test_size=0.3, random_state=1)
```

```python
# выполняем дамми-кодирование
clf_train_data = pd.get_dummies(clf_train_data)
clf_test_data = pd.get_dummies(clf_test_data)
```

```python
# смотрим форму каждого массива
print(clf_train_data.shape)
print(clf_train_labels.shape)
print(clf_test_data.shape)
print(clf_test_labels.shape)
```

```python
# переводим в массивы NumPy
train_data = clf_train_data.values
test_data = clf_test_data.values
train_labels = clf_train_labels.values.ravel()
test_labels = clf_test_labels.values.ravel()
```

```python
# импортируем класс XGBClassifier и функцию roc_auc_score
from xgboost import XGBClassifier
from sklearn.metrics import roc_auc_score
```

```python
# строим модель
xgb = XGBClassifier(n_estimators=150, learning_rate=0.01, 
                    max_depth=50, subsample=0.9, 
                    colsample_bytree=1, colsample_bylevel=0.9, 
                    reg_alpha=0.1, reg_lambda=0.1, random_state=0, n_jobs=-1, verbose=1)
xgb.fit(train_data, train_labels)
# печатаем значения правильности
print("AUC на обучающей выборке: {:.3f}".
      format(roc_auc_score(train_labels, xgb.predict_proba(train_data)[:, 1])))
print("AUC на контрольной выборке: {:.3f}".
      format(roc_auc_score(test_labels, xgb.predict_proba(test_data)[:, 1])))
```
